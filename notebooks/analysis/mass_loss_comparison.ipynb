```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np

proj_dir = os.path.join(os.pardir,os.pardir)
data_dir = os.path.join(proj_dir,'data')
fig_dir = os.path.join(proj_dir,'reports','figures')
```

```python
sheet_url = ('https://docs.google.com/spreadsheets/d/1soBv6c4haG7xK3sz4Mbxsl-BKYjOAoZE5lxvDk0hetU/edit?usp=sharing')
csv_export_url = sheet_url.replace('edit?', 'export?gid=0&format=csv&')
mlc = pd.read_csv(csv_export_url,index_col=0,na_values=['?','-',' '])
mlc
```

```python
out_path = os.path.join(fig_dir,'bmb_obs.png')

data = mlc.iloc[0:1][['Liu 2015 (non steady)','Liu 2015 (steady state)','Depoorter 2013','Rignot 2013','Adusumilli 2018']]
errors = mlc.iloc[0:1][['Liu 2015 err (non steady)','Liu 2015 err (steady state)','Depoorter 2013 err','Rignot 2013 err']]
errors.columns=['Liu 2015 (non steady)','Liu 2015 (steady state)','Depoorter 2013','Rignot 2013']

plt.close()
fig,ax = plt.subplots(figsize=(15,5))
mlc.iloc[0:1].plot(kind='barh',y='This study (all ice)',ax=ax,color='silver',width=1,fontsize=18)
data.plot(kind='barh',xerr=errors,ax=ax,width=0.7,alpha=0.7)
ax.legend(markerscale=1,fontsize=18,bbox_to_anchor=(1.,1.1))
ax.set_title('Antarctic mass loss compared against observations',fontsize=24)
ax.set_xlabel('Mass loss in Gt/yr',fontsize=18)
ax.set_yticklabels("")
ax.grid(axis='x')
plt.savefig(out_path,dpi=300,format='png',bbox_to_anchor=True)
plt.show()
```
